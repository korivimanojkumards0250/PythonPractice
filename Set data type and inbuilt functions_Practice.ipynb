# Set data type and inbuilt functions
###set does not holds any index numbers

### set.add(value) to add a single element to the current set
### set.copy(newSetName) duplicates a set
### set.clear()   removes all the elements in the set
### set.pop()  removes random element in the set
### set.remove(value)  This removed value and throws error if value not found
### set.discard(value)   This removes value if exists in the set otherwise ignores it
### set1.union(set2)  It will join two sets and stores in new memory location(original set remains same)
### set1.update(set2) It will join two sets but stores in same original memory location (original set will update with new elements)
### set1.intersection(set2)   it will find the common elements and store in seperate memory
### set1.intersection_update(set2)  it will find the common elements and updates the calling set with the results
### set1.difference(set2)    it will find the remaining elements other than the common elements in the calling set and keeps in new memory location.
### set1.difference_update(set2)   it will find the remaining elements other than the common elements and updates to the calling set.
### set1.symmetric_difference(set2)  it will find the elements which are not common in both the sets and assign to a new memory location
### set1.symmetric_difference_update(set2)   it will find the elements which are not common in both the sets and updates as elements in the calling set
### issubset  it will check if a set is subset of another set and gives results as either true or false
### issuperset  its a opposite of subset
### isdisjoint  it will gives true if there are no common elements in both the sets

```python
set={1,1,2,3,4}
type(set)
set
```

```python
set1={1,2,3,4}
set2={4,5,6,7}
set1.add(7)
print(set1)
```

```python
set1={1,2,3,4}
set2={3,4,5,6,7}
a=set1.union(set2)
print(set1)
print(a)
a=set1.intersection(set2)
print(a)
print(set1)
a=set1.update(set2)
print(set1)
print(a)
```

```python
set1={1,2,3,4}
set2={3,4,5,6,7}
a=set1.intersection(set2)
print(a)
print(set1)
print(set2)
#variable a is allocated with new memory location with common elements in both sets
a=set1.intersection_update(set2)
print(a)
print(set1)
print(set2)
```

```python
set1={1,2,3,4}
set2={3,4,5,6,7}
a=set1.difference(set2)
print(a)
print(set1)
a=set1.difference_update(set2)
print(a)
print(set1)
print(set2)
```

```python
set1={1,2,3,4}
set2={3,4,5,6,7}
a=set1.symmetric_difference(set2)
b=set2.symmetric_difference(set1)
print(a)
print(b)
print(set1)
a=set1.symmetric_difference_update(set2)
print(a)
print(set1)
print(set2)
```

```python
set1={1,2,3,4}
set2={1,2,3,4,5,6,7}
set1.issubset(set2)
```

```python
set1={1,2,3,4,5,6,7,8}
set2={1,2,3,4,5,6,7}
set1.issuperset(set2)
```

```python
set1={1,2,3,4}
set2={5,6,7}
set1.isdisjoint(set2)
```
