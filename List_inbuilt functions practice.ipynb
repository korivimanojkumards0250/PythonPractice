# List inbuilt functions
### List.append(values) add values of any data type to the current existing list
### List.extend(listName) to add extra values within the same list
### List.count(value)   to check the total count of particular value
### List.index(value)   to check the index number of particular value
### List.insert(index, value)   to insert the value at particular index location
### List.remove(value)    to remove any particular value(only first instance)
### List.reverse()    to reverse the values in the list
### List.sort()       to sort the values in ascending order(list with only numbers or float values will work to sort)
### List.sort(reverse=True)    to sort the values in descending order. reverse=False for ascending order

```python
d=[1,2,3,4,5,6]
print(d)
##List inbuild functions
# list.append()   can add values of any data type to the current existing list
#to add another list to the exsting list
d.append([4,34])
print(d)
#append extra values to the same list
d.append(23)
d.append("Manoj")
print(d)
# list.extend() to add extra values within the same list
d.extend([22,11])
#even though we gave it as list, it just added as seperate values - not a list
print(d)
```

```python
#Shallow copy Vs Deep copy
#Shallow copy used the same memory location - so it chnages the original values if we update anything
L4=[4,5,6,7]
L5=L4
L5[2]=33
print(L4,L5)
```

```python
#deep copy creates a copy of the variable so that original variable will not change its values
L4=[4,5,6,7]
L5=L4.copy()
L5[2]=33
print(L4,L5)
```
